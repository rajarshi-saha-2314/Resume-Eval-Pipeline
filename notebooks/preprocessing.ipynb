```python
import pandas as pd
import re
```

```python
# Load dataset
df = pd.read_csv("../data/raw/resumes.csv")
```

```python
# Clean column names
df.columns = df.columns.str.strip()
```

```python
# Rename columns
df = df.rename(columns={
    "Resume_str": "text",
    "Category": "label"
})
```

```python
# Keep only needed columns
df = df[["text", "label"]].copy()
```

```python
# Drop missing rows
df = df.dropna(subset=["text", "label"])
```

```python
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"[^a-zA-Z\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text
```

```python
# Apply preprocessing
df["clean_text"] = df["text"].apply(clean_text)
```

```python
# Final processed dataframe
processed_df = df[["clean_text", "label"]]
```

```python
# Save processed data
processed_df.to_csv("../data/processed/processed_resumes.csv", index=False)
```
