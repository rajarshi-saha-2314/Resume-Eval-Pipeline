```python
import pandas as pd
import pickle
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# Load processed data
df = pd.read_csv("../data/processed/processed_resumes.csv")
df = df.dropna(subset=["clean_text", "label"])

X = df["clean_text"]
y = df["label"]
```

```python
# Load Logistic Regression model
with open("../models/logistic_regression/model.pkl", "rb") as f:
    lr_model, lr_vectorizer = pickle.load(f)
```

```python
# Load Random Forest model
with open("../models/random_forest/model.pkl", "rb") as f:
    rf_model, rf_vectorizer = pickle.load(f)
```

```python
# Transform inputs
X_lr = lr_vectorizer.transform(X)
X_rf = rf_vectorizer.transform(X)
```

```python
# Predictions
lr_preds = lr_model.predict(X_lr)
rf_preds = rf_model.predict(X_rf)
```

```python
# Compare metrics
results = pd.DataFrame([
    {
        "Model": "Logistic Regression",
        "Accuracy": accuracy_score(y, lr_preds),
        "Precision": precision_score(y, lr_preds, average="weighted"),
        "Recall": recall_score(y, lr_preds, average="weighted"),
        "F1 Score": f1_score(y, lr_preds, average="weighted")
    },
    {
        "Model": "Random Forest",
        "Accuracy": accuracy_score(y, rf_preds),
        "Precision": precision_score(y, rf_preds, average="weighted"),
        "Recall": recall_score(y, rf_preds, average="weighted"),
        "F1 Score": f1_score(y, rf_preds, average="weighted")
    }
])
```

```python
print(results)
```
